```python
import pandas as pd

# Load dataset
df = pd.read_csv('ground_vehicles.csv')
```

```python
df
```

```python
# Load the standardized vendor name mapping table
mapping = pd.read_csv('mapping.csv')
```

```python
mapping
```

```python
# Merge the dataset with the mapping table to replace variations
normalized_df = pd.merge(df, mapping, left_on='Vendor  Name', right_on='Standardized Name', how='right')
```

```python
normalized_df
```

```python
# Rename the standardized "New Vendor Name" column, and drop the 'Variation' 'Vendor Name', &'Standardized Name' columns
normalized_df['New Vendor Name'] = normalized_df['Standardized Name']
normalized_df.drop(['Variation', 'Vendor  Name','Standardized Name'], axis=1, inplace=True)
```

```python
# Save the updated dataset
normalized_df.to_csv('cleaned_data.csv', index=False)
```

```python
#Group by Fiscal Year and add total Awarded Amount
spending_by_year = data.groupby('Fiscal Year')['Awarded Amount'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Create a line plot to visualize spending trends
plt.figure(figsize=(10, 5))
plt.plot(spending_by_year['Fiscal Year'], spending_by_year['Awarded Amount'], marker='o')
plt.xlabel('Fiscal Year')
plt.ylabel('Total Spending')
plt.title('Spending Trends Over Fiscal Years')
plt.grid(True)
plt.show()
```

```python
# Calculate percent increase over time
spending_by_year['Percent Increase'] = spending_by_year['Awarded Amount'].pct_change() * 100

# Create a line plot to visualize percent increases with colored lines
plt.figure(figsize=(10, 5))
for i in range(len(spending_by_year) - 1):
    percent_increase = spending_by_year['Percent Increase'].iloc[i:i+2]
    color = 'green' if percent_increase.iloc[1] > 0 else 'red'
    plt.plot(spending_by_year['Fiscal Year'].iloc[i:i+2], percent_increase, marker='o', color=color, linestyle='-')

plt.xlabel('Fiscal Year')
plt.ylabel('Percent Increase (%)')
plt.title('Percent Increase in Awarded Amount Over Fiscal Years')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(spending_by_year['Fiscal Year'])
plt.axhline(0, color='gray', linewidth=0.5)  # Add a horizontal line at y=0
plt.tight_layout()

# Annotate the data points with percent increase values
for x, y, color in zip(spending_by_year['Fiscal Year'], spending_by_year['Percent Increase'], ['green' if val > 0 else 'red' for val in spending_by_year['Percent Increase']]):
    plt.annotate(f'{y:.2f}%', (x, y), textcoords="offset points", xytext=(0, 10), ha='center', color=color)

legend_labels = ['Increasing', 'Decreasing']
legend_handles = [plt.Line2D([0], [0], color='g', lw=2), plt.Line2D([0], [0], color='r', lw=2)]

# Define legend_handles before using it in the legend
plt.legend(legend_handles, legend_labels, loc='upper left', title='Trend')

plt.show()
```

```python
# Define a function to categorize titles
def categorize_title(title):
    if pd.isna(title):
        return 'Null'
    elif 'M1A1' in title or 'ABRAMS TANK' in title or 'ABRAMS TANKS' in title:
        return 'Abrams Tank'
    elif 'M2A1' in title or 'BRADLEY FIGHTING VEHICLE'in title:
        return 'Bradley Fighting Vehicle'
    elif 'M1130' in title or 'STRYKER ARMORED PERSONNEL CARRIER'in title:
        return 'Stryker Armored Personnel Carrier'
    else:
        return 'Other'

# Apply the function to the Title column to create a new 'Category' column
data['Category'] = data['Title'].apply(categorize_title)

# Count the occurrences of each category in the entire dataset
category_counts = data['Category'].value_counts()

# Display the result
print(category_counts)
```

```python
# Filter out the "Other" category
category_counts_wo_other = category_counts[category_counts.index != 'Other']

# Create a pie chart
plt.pie(category_counts_wo_other, labels=category_counts_wo_other.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Total Awarded Amount by GCS Program (excluding "Other")')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
spending_by_program = data.groupby('Category')['Awarded Amount'].sum().reset_index()

# Sort the data by 'Awarded Amount' in ascending order
spending_by_program = spending_by_program.sort_values(by='Awarded Amount', ascending=True)
```

```python
import matplotlib.colors as mcolors
from matplotlib.cm import ScalarMappable

normalized_data = (spending_by_program['Awarded Amount'] - spending_by_program['Awarded Amount'].min()) / (
    spending_by_program['Awarded Amount'].max() - spending_by_program['Awarded Amount'].min()
)

#Define a custom colormap from green to yellow to red
colors = [(0, 'red'), (0.01, 'yellow'), (1, 'green')]
custom_cmap = mcolors.LinearSegmentedColormap.from_list('custom_colormap', colors)

# Create a ScalarMappable object to map normalized values to colors
sm = ScalarMappable(cmap=custom_cmap, norm=plt.Normalize(0, 1))

# Get the RGBA values for the colors based on normalized data
colors = sm.to_rgba(normalized_data)

# Create a bar chart with the custom gradient colors
plt.figure(figsize=(12, 6))
plt.bar(spending_by_program['Category'], spending_by_program['Awarded Amount'], color=colors)
plt.xlabel('GCS Programs')
plt.ylabel('Total Awarded Amount (log scale)')
plt.yscale('log')  # Set the y-axis scale to logarithmic
plt.title('Total Awarded Amount by GCS Program')

legend_labels = ['High', 'Medium','Low']
legend_handles = [plt.Line2D([0], [0], color='g', lw=3), plt.Line2D([0], [0], color='orange', lw=3), plt.Line2D([0], [0], color='r', lw=3)]

# Define legend_handles before using it in the legend
plt.legend(legend_handles, legend_labels, loc='upper left', title='')

# Display the bar chart
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group the original data by 'Fiscal Year' and 'Category', and sum 'Awarded Amount'
spending_by_year_category = data.groupby(['Fiscal Year', 'Category'])['Awarded Amount'].sum().reset_index()

# Filter the data to include categories
categories_to_include = ['Abrams Tank', 'Stryker Armored Personnel Carrier','Bradley Fighting Vehicle', 'Other', 'Null']
filtered_data = spending_by_year_category[spending_by_year_category['Category'].isin(categories_to_include)]

# Create a line plot to visualize spending trends for the filtered data
plt.figure(figsize=(10, 5))
for category in categories_to_include:
    category_data = filtered_data[filtered_data['Category'] == category]
    plt.plot(category_data['Fiscal Year'], category_data['Awarded Amount'], marker='o', label=category)

plt.xlabel('Fiscal Year')
plt.ylabel('Total Spending')
plt.title('Spending Trends Over Fiscal Years Per GCS Program')
plt.legend()
plt.yscale('log')
plt.grid(True, linestyle='--', alpha=0.7)
# Put a legend to the right of the current axis
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Filter out "Other" values in the 'Category' column
data_filtered = data[data['Category'] != 'Other']

# Group the data by 'Vendor Name' and 'Category', and sum 'Awarded Amount'
spending_by_vendor_category = data_filtered.groupby(['Category', 'New Vendor Name'])['Awarded Amount'].sum().reset_index()

# Sort vendors by total awarded amount and select the top 20
top_vendors = spending_by_vendor_category.groupby('New Vendor Name')['Awarded Amount'].sum().nlargest(20).index

# Create a custom color palette with a sufficient number of distinct colors
custom_palette = plt.cm.get_cmap('tab20', len(top_vendors))

plt.figure(figsize=(12, 10)) 

for i, vendor in enumerate(top_vendors):
    vendor_data = spending_by_vendor_category[spending_by_vendor_category['New Vendor Name'] == vendor]
    plt.bar(vendor_data['Category'], vendor_data['Awarded Amount'], label=f'{vendor}', color=custom_palette(i))

plt.xlabel('GCS Program')
plt.ylabel('Total Amount')
plt.title('Awarded Amount by Category and their Top 20 Vendors')
plt.yscale('log')
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.7)

plt.legend()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
